<p>Since the large datasets available, I've decided to use R to prepared and modify the data.

```r
#---------First let's import the dplyr library and the stations dataset.------------------------------
library(tidyverse)
library(lubridate)
stations2017 = read.csv('../input/stations2017/Divvy_Stations_2017_Q3Q4.csv')

#---------Let's now change the varibles types to the appropiate format and delete one column ---------
stations2017$online_date = mdy_hm(stations2017$online_date)
stations2017$X = NULL

#---------Let's now run a check for duplicates or errors-----------------------------------------------
print(duplicated(stations2017))
print(is.null(stations2017))
stations2017 = stations2017 %>% mutate(year = year(online_date))
stations2017
```

```r
#---------Now let's plot how the company grew overtime by plotting the stations in the map------------
library(openintro)
library(ggmap)
library(ggthemes)
library(maps)

chicago_map = get_stamenmap(
        bbox = c(left = -88.3177, bottom =41.4561, right =-86.8208, top =42.1461),
        maptype = "terrain",
        zoom = 10)

ggmap(chicago_map) +
  geom_point(data = stations2017, 
            aes(x = longitude, y = latitude, color = , alpha = 0.5),
            size = 0.5) + 
  theme_map()
```
